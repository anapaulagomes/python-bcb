```python
url = "https://olinda.bcb.gov.br/olinda/servico/selic_operacoes/versao/v1/odata/"
service = ODataAPI(url)
service.describe()
```

```python
service.describe("Selic")
```

```python
ep = service.get_endpoint('Selic')
ep.query().limit(10).collect()
```

```python
df = ep.query().filter(ep.anoMes >= "2020-00").collect()
```

```python
df.groupby("Modalidade").agg({"TaxaJurosAoAno": "median"})
```

```python
df.groupby(["Modalidade", "anoMes"]).agg({"TaxaJurosAoAno": "median"})
```
